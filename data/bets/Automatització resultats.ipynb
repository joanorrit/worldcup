```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

# Desactivar Arial completamente
mpl.rcParams['font.sans-serif'] = ['Segoe UI Emoji']
mpl.rcParams['font.family'] = 'sans-serif'

# Evitar que Matplotlib intente usar Arial como fallback
mpl.rcParams['font.fantasy'] = ['Segoe UI Emoji']
mpl.rcParams['font.serif'] = ['Segoe UI Emoji']
mpl.rcParams['font.monospace'] = ['Segoe UI Emoji']
mpl.rcParams['font.cursive'] = ['Segoe UI Emoji']

# Estilo visual moderno
sns.set_theme(style="whitegrid")
palette = sns.color_palette("viridis", 10)

def plot_histograms(df, columns, yticks, title, filename, height=4):
    """
    Crea una columna de histogramas con estilo profesional y los guarda en PNG.
    """
    fig, axes = plt.subplots(len(columns), 1, figsize=(12, height * len(columns)))

    if len(columns) == 1:
        axes = [axes]

    for i, col in enumerate(columns):
        sns.histplot(
            df[col],
            ax=axes[i],
            kde=False,
            color=palette[i % len(palette)],
            edgecolor="black",
            linewidth=1.2
        )
        axes[i].set_title(col, fontsize=16, fontweight="bold")
        axes[i].set_ylabel("Freqüència", fontsize=12)
        axes[i].set_xlabel("")
        axes[i].set_yticks(yticks)
        axes[i].grid(axis="y", linestyle="--", alpha=0.4)

        # Nombres en diagonal
        axes[i].tick_params(axis='x', rotation=45)

    fig.suptitle(title, fontsize=20, fontweight="bold", y=1.02)

    plt.tight_layout()

    # Guardar en PNG
    fig.savefig(filename, dpi=300, bbox_inches="tight")

    plt.show()


# -----------------------------
#   GRUP 1: Podis
# -----------------------------
plot_histograms(
    res,
    ["Campió", "Subcampió", "3er Lloc"],
    yticks=list(range(1, 14)),
    title="Podis del Torneig",
    filename=ABS_PATH + "Figures/Podis.png"
)

# -----------------------------
#   GRUP 2: Botes
# -----------------------------
plot_histograms(
    res,
    ["Bota Or", "Bota Plata", "Bota Bronze"],
    yticks=list(range(1, 10)),
    title="Botes del Torneig",
    filename=ABS_PATH + "Figures/Botes.png"
)

# -----------------------------
#   GRUP 3: Millor Jugador
# -----------------------------
plot_histograms(
    res,
    ["Millor Jugador"],
    yticks=list(range(1, 10)),
    title="Millor Jugador",
    filename=ABS_PATH + "Figures/Millor_Jugador.png"
)

# -----------------------------
#   GRUP 4: 2n i 3r Millor Jugador
# -----------------------------
plot_histograms(
    res,
    ["2n Millor Jugador", "3er Millor Jugador"],
    yticks=list(range(1, 5)),
    title="Millors Jugadors",
    filename=ABS_PATH + "Figures/Millors_Jugadors.png"
)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Asegurar índice datetime
df_t.index = pd.to_datetime(df_t.index)
df_t = df_t.sort_index()

# Número de frames entre días
FRAMES_PER_DAY = 100

# Crear un índice temporal interpolado
full_index = pd.date_range(
    start=df_t.index.min(),
    end=df_t.index.max(),
    periods=(len(df_t)-1)*FRAMES_PER_DAY + 1
)
# Eliminar fechas duplicadas en el índice
df_t.index = pd.to_datetime(df_t.index)
df_t = df_t[~df_t.index.duplicated(keep="last")]
df_t = df_t.sort_index()

# Interpolar valores suavemente
df_interp = df_t.reindex(df_t.index.union(full_index))
df_interp = df_interp.interpolate(method="linear").reindex(full_index)

# Configuración de estilo
plt.rcParams["font.family"] = "Segoe UI Emoji"
fig, ax = plt.subplots(figsize=(12, 7))

def update(frame):
    ax.clear()

    fecha = df_interp.index[frame]
    datos = df_interp.loc[fecha].dropna().sort_values(ascending=True)

    # Dibujar barras ordenadas
    ax.barh(datos.index, datos.values, color="royalblue")

    # Título dinámico
    ax.set_title(f"🏆 Evolució de puntuacions — {fecha.date()}", fontsize=20, fontweight="bold")

    # Etiquetas
    ax.set_xlabel("Punts", fontsize=14)
    ax.set_xlim(0, df_interp.max().max() * 1.15)

    # Valores al final de cada barra
    for i, (jug, val) in enumerate(zip(datos.index, datos.values)):
        ax.text(val + 2, i, f"{val:.0f}", va="center", fontsize=11)

    # Ranking dinámico
    for i, jug in enumerate(datos.index):
        ax.text(0, i, f"{i+1}º", va="center", fontsize=11, fontweight="bold", color="gray")

    plt.tight_layout()

anim = FuncAnimation(
    fig,
    update,
    frames=len(df_interp),
    interval=60,   # velocidad (ms por frame)
    repeat=True
)

plt.show()
from matplotlib.animation import FFMpegWriter

writer = FFMpegWriter(fps=30, bitrate=1800)

anim.save("race_suave.mp4", writer=writer, dpi=200)
```

```python
import matplotlib.pyplot as plt 
import numpy as np

colors = plt.cm.jet(np.linspace(0,1,13))

def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=8, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]

def prepare_data(df, steps=8):
    df = df.reset_index()
    df.index = df.index * steps
    last_idx = df.index[-1] + 1
    df_expanded = df.reindex(range(last_idx))
    df_expanded['index'] = df_expanded['index'].fillna(method='ffill')
    df_expanded = df_expanded.set_index('index')
    df_rank_expanded = df_expanded.rank(axis=1, method='first')
    df_expanded = df_expanded.interpolate()
    df_rank_expanded = df_rank_expanded.interpolate()
    return df_expanded, df_rank_expanded

df_expanded, df_rank_expanded = prepare_data(df_t)
df_expanded.head()

from matplotlib.animation import FuncAnimation

def init():
    ax.clear()
    nice_axes(ax)
    #ax.set_ylim(.2, 6.8)

def update(i):
    for bar in ax.containers:
        bar.remove()
    y = df_rank_expanded.iloc[i]
    width = df_expanded.iloc[i]
    labels = df_expanded.columns
    ax.barh(y=y, width=width, color=colors, tick_label=labels)
    
    date_str = df_expanded.index[i].strftime('%Y-%m-%d')

    if date_str < '2024-06-29':
        text = '(Fase de Grups)'
    else:
        if date_str < '2024-07-03':
            text = '(Vuitens de Final)'
        elif date_str < '2024-07-09':
            text = '(Quarts de final)'
        elif date_str < '2024-07-12':
            text = '(Semifinal)'       
        else:
            text = '(Final)'

    
    #date_str = "Avui"
    #text = "Avui"
    ax.set_title(f"Classificació Euro '24 - {date_str} {text}", fontsize='smaller')
    ax.set_xlabel('Punts')

    
fig = plt.Figure(figsize=(5, 2.5), dpi=150)
ax = fig.add_subplot()
anim = FuncAnimation(fig=fig, func=update, init_func=init, frames=len(df_expanded), 
                     interval=150, repeat=False)
```
